For a baseline model, we're going to recommend only the top 25 songs of the past year for all people, regardless of listening history.

To do this, we'll query the Metalistenz database for this information.

```python
# practice
%%bigquery --project listenbrainzrecommender practice
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.samples.gsod`
 
```

```python
%%bigquery --project listenbrainzrecommender top25songs_alltime
SELECT
  recording_mbid,
  COUNT(user_name) AS song_plays

FROM
  `listenbrainz.listenbrainz.listen`

WHERE
  LENGTH(recording_mbid) > 0

GROUP BY
  recording_mbid

ORDER BY
  song_plays DESC

LIMIT
  25
```
